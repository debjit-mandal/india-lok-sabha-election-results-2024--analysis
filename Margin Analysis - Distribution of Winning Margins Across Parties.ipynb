
# Margin Analysis - Distribution of Winning Margins Across Parties

This analysis examines the distribution of winning margins across different political parties in the 2024 Lok Sabha Elections using visualizations.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/election_results_2024.csv')
data['Margin'] = pd.to_numeric(data['Margin'], errors='coerce')

data_cleaned = data.dropna(subset=['Margin'])
```

## Visualisation of Distribution of Winning Margins


```python
plt.figure(figsize=(12, 8))
sns.boxplot(y='Leading Party', x='Margin', data=data_cleaned)
plt.xticks(rotation=90)
plt.title('Distribution of Winning Margins Across Different Parties')
plt.ylabel('Leading Party')
plt.xlabel('Winning Margin')
plt.tight_layout()
plt.show()
```


## Conclusion
The above analysis provides insights into how close or decisive the victories were for each party in the 2024 Lok Sabha Elections.
