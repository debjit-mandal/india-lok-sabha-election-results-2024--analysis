
# Party Performance - Total Seats and Vote Share

This analysis calculates the total number of seats won by each party and determines their vote share percentage in the 2024 Lok Sabha Elections.


```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('data/election_results_2024.csv')
data['Margin'] = pd.to_numeric(data['Margin'], errors='coerce')
```

## Calculation of Winning Seats


```python
seats_won = data['Leading Party'].value_counts().reset_index()
seats_won.columns = ['Party', 'Seats Won']
```


## Calculation of Vote Share


```python
total_votes = data['Margin'].sum()
party_votes = data.groupby('Leading Party')['Margin'].sum().reset_index()
party_votes['Vote Share (%)'] = (party_votes['Margin'] / total_votes) * 100
```

## Results Merged


```python
party_performance = pd.merge(seats_won, party_votes, left_on='Party', right_on='Leading Party')
party_performance = party_performance[['Party', 'Seats Won', 'Vote Share (%)']]
party_performance
```

# Visualization of Party Performance


## Bar plot for seats won by each party


```python
plt.figure(figsize=(12, 12))
plt.barh(party_performance['Party'], party_performance['Seats Won'], color='blue')
plt.ylabel('Party')
plt.xlabel('Seats Won')
plt.title('Total Seats Won by Each Party')
plt.gca().invert_yaxis()
plt.show()
```

## Bar Plot for vote share percentage of each party

```python
plt.figure(figsize=(12, 12))
plt.barh(party_performance['Party'], party_performance['Vote Share (%)'], color='green')
plt.ylabel('Party')
plt.xlabel('Vote Share (%)')
plt.title('Vote Share Percentage of Each Party')
plt.gca().invert_yaxis()
plt.show()
```

## Conclusion
The above analysis provides insights into the performance of each party in terms of seats won and vote share in the 2024 Lok Sabha Elections.
