
# Margin Analysis - High and Low Winning Margins

This analysis identifies the constituencies with the highest and lowest winning margins in the 2024 Lok Sabha Elections.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('data/election_results_2024.csv')
```

## Data Preprocessing


```python
data['Margin'] = pd.to_numeric(data['Margin'], errors='coerce')
```


## Highest Winning Margins


```python
highest_margins = data.nlargest(10, 'Margin')[['Constituency', 'Leading Candidate', 'Margin', 'Leading Party']]
highest_margins
```

## Visualization of Highest Winning Margins


```python
plt.figure(figsize=(12, 6))
plt.barh(highest_margins['Constituency'], highest_margins['Margin'], color='green')
plt.xlabel('Winning Margin')
plt.ylabel('Constituency')
plt.title('Top 10 Constituencies with Highest Winning Margins')
plt.gca().invert_yaxis()
plt.show()
```


## Lowest Winning Margins


```python
lowest_margins = data.nsmallest(10, 'Margin')[['Constituency', 'Leading Candidate', 'Margin', 'Leading Party']]
lowest_margins
```

## Visualization of Lowest Winning Margins


```python
plt.figure(figsize=(12, 6))
plt.barh(lowest_margins['Constituency'], lowest_margins['Margin'], color='red')
plt.xlabel('Winning Margin')
plt.ylabel('Constituency')
plt.title('Top 10 Constituencies with Lowest Winning Margins')
plt.gca().invert_yaxis()
plt.show()
```


## Conclusion
The above analysis provides insights into the constituencies with the highest and lowest winning margins in the 2024 Lok Sabha Elections.
