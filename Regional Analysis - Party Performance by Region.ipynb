
# Regional Analysis - Party Performance by Region

This analysis examines the performance of different political parties across various regions in the 2024 Lok Sabha Elections.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/election_results_2024.csv')
data['Margin'] = pd.to_numeric(data['Margin'], errors='coerce')
```

## Addding State Information
For this analysis, I am adding a hypothetical 'State' column to the dataset.

```python
np.random.seed(0)
states = ['State1', 'State2', 'State3', 'State4', 'State5']
data['State'] = np.random.choice(states, len(data))
```

## Group Data by State and Part

```python
regional_performance = data.groupby(['State', 'Leading Party']).size().reset_index(name='Seats Won')
regional_performance_pivot = regional_performance.pivot(index='State', columns='Leading Party', values='Seats Won').fillna(0)
regional_performance_pivot
```

## Visualization of Regional Performance


```python
plt.figure(figsize=(12, 8))
sns.heatmap(regional_performance_pivot, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Regional Performance of Parties')
plt.xlabel('Leading Party')
plt.ylabel('State')
plt.show()
```

## Conclusion
The above analysis provides insights into the performance of different parties in various regions in the 2024 Lok Sabha Elections.
